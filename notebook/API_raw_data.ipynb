```python
#Import library
import requests
import pandas as pd
import sqlite3
```

```python
# =====================
# CONFIG
# =====================

API_KEY = "Your API Key here"
BASE_URL = "https://financialmodelingprep.com/stable"

SYMBOLS = ["AMZN", "GOOGL"]
MARKET_INDEX = "^IXIC"

START_DATE = "2024-09-24"
END_DATE = "2025-10-06"
```

```python
#Get income statement
def get_quarterly_income(symbol):
    url = (
        f"{BASE_URL}/income-statement"
        f"?symbol={symbol}"
        f"&period=quarter"
        f"&limit=5"
        f"&apikey={API_KEY}"
    )

    response = requests.get(url)
    response.raise_for_status()

    data = response.json()
    df = pd.DataFrame(data)
    df["symbol"] = symbol

    return df
```

```python
#Get stock price
def get_eod_full_prices(symbol):
    url = (
        f"{BASE_URL}/historical-price-eod/full"
        f"?symbol={symbol}"
        f"&from={START_DATE}"
        f"&to={END_DATE}"
        f"&apikey={API_KEY}"
    )

    response = requests.get(url)
    response.raise_for_status()

    data = response.json()
    df = pd.DataFrame(data)
    df["symbol"] = symbol

    return df
```

```python
#Income statement to dataframe
earnings_df = pd.concat(
    [get_quarterly_income(symbol) for symbol in SYMBOLS],
    ignore_index=True
)

earnings_df
```

```python
#Stock price to dataframe
prices_df = pd.concat(
    [get_eod_full_prices(symbol) for symbol in SYMBOLS],
    ignore_index=True
)

prices_df.head()
```

```python
#Market price to dataframe
market_df = get_eod_full_prices(MARKET_INDEX)
market_df.rename(columns={"symbol": "index_symbol"}, inplace=True)

market_df.head()
```

```python
#Export to csv
earnings_df.to_csv("raw_quarterly_earnings.csv", index=False)
prices_df.to_csv("raw_daily_stock_prices.csv", index=False)
market_df.to_csv("raw_market_index_prices.csv", index=False)
```
